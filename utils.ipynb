```python
from pathlib import Path
import shutil as sh
from bs4 import BeautifulSoup

def list_pages():
    import glob
    pages = glob.glob('*.html')
    pages.remove("index.html")
    return pages

def save(page:Path, soup:BeautifulSoup):
    page.write_text(soup.prettify())

def backup(page:Path, folder=None):
    backup = page.absolute().with_name(page.stem+"-backup").with_suffix(page.suffix)
    sh.copyfile(page, backup) 

list_pages()
```

## Set default attributes on videotags in index.html
`<video controls loop muted poster="{original}"></video>`

```python
from pathlib import Path
index = Path("./index.html")


from bs4 import BeautifulSoup
def set_all_video_element_default_attributes(soup):
    for video in soup.find_all("video"):
        poster = video.attrs['poster']
        video.attrs = {
            'loop': None,
            'muted': None,
            'controls': None,
            'poster': poster,
        }

soup = BeautifulSoup(index.read_text())
set_all_video_element_default_attributes(soup)

for video in soup.find_all("video"):
    print(video.attrs)

# save(index, soup)
```
